```python
from syntax_analyzer import SyntaxAnalyzer
```

```python
import pandas as pd
data = pd.read_csv('./data.csv', encoding='windows-1251') #data from 01.01.2022 to 10.12.2023
```

```python
from tqdm import tqdm
analyzer = SyntaxAnalyzer()
sentences = []
for text in tqdm(data['text'][14500:].tolist()):
    doc = analyzer(text)

    sentences.append(analyzer.get_sentences(doc, normalize=True, upos=['NOUN', 'VERB', 'AUX', 'ADJ', 'ADV', 'X']))
```

```python
sr = pd.Series(sentences)
```

```python
sr.to_csv('./normalized11-13.csv')
```
